```python
import pandas as pd
import numpy as np
data=pd.read_csv('final_vg[1].csv')
data.head()
```

```python
# Overall count of Genre column
data['Genre'].count()
```

```python
# Count of each Genre
cat_counts=data['Genre'].value_counts()
cat_counts
```

```python
x_bar=cat_counts.index
y_bar=cat_counts
plt.bar(x_bar,y_bar)
```

```python
# Chose initial 5 values
plt.bar(x_bar[:5],y_bar[:5])
```

```python
# Resize or increase the size of fig
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar)
```

```python
# Rotate the ticks
plt.bar(x_bar,y_bar)
plt.xticks(rotation=90)
```

```python
 #labelling
plt.bar(x_bar[:5],y_bar[:5])
plt.xlabel("Genre")
plt.ylabel("Count")
```

```python
# Changing color
plt.xlabel("Genre")
plt.ylabel("Count")
plt.bar(x_bar[:5],y_bar[:5],color='orange')
plt.title("Distribution of Video Game")
```

```python
# Changing width and color
# Changing color
plt.xlabel("Genre")
plt.ylabel("Count",fontsize=20)
plt.bar(x_bar[:5],y_bar[:5],color='g',width =0.3)
plt.title("Distribution of Video Game")
```

```python
plt.figure(figsize=(15,8))
plt.bar(x_bar,y_bar,color='olive',width=0.5)
plt.xticks(rotation=45)
plt.xlabel("Genre",fontsize=14)
plt.ylabel("Count",fontsize=18)
plt.title("Distribution of videogame by Genre")
```

```python
# plt.show() make sure only the final chart will display 
# it ensure that upcoming chart is not getting change
# use all changes in single cell
```

```python
plt.figure(figsize=(15,8))
plt.bar(x_bar,y_bar,color='olive',width=0.5)
plt.xticks(rotation=45)
plt.xlabel("Genre",fontsize=14)
plt.ylabel("Count",fontsize=18)
plt.title("Distribution of videogame by Genre")
plt.show()
```
