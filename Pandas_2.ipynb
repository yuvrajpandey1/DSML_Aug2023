#### Merge ,Join

```python
users = pd.DataFrame({'userid':[1,2,3],"name":["Yuvraj","Gitansh","Raghav"]})
users
```

```python
msgs = pd.DataFrame({'userid':[1,1,2,4],"msg":['hmm','acha','theek hai','nice']})
msgs
```

```python
pd.concat([users,msgs])
```

```python
pd.concat([users,msgs],axis=1)
# Why it showing float
# Nan is consider as flot value
```

```python
users
```

```python
msgs
```

```python
# need to give common col 
users.merge(msgs,on='userid')  # inner join
```

```python
# common value merger other enhanced
users.merge(msgs,on='userid',how='outer')  # outer join
```

```python
users.merge(msgs,on='userid',how='left')  # left join
```

```python
users.merge(msgs,on='userid',how='right')  # right join
```

```python
users.rename({'userid':'user'},axis=1)
```

```python
users.rename(columns={'userid':'user'},inplace=True)
users
```

```python
users.merge(msgs,left_on='user',right_on='userid')  # inner join
```
