```python
# reduce data for
# top3 publisher
# top3 plateform
# top3 genre
```

```python
top3_publisher=data['Publisher'].value_counts().index[:3]
top3_plateform=data['Platform'].value_counts().index[:3]
top3_genre=data['Genre'].value_counts().index[:3]
```

```python
top3_data=data.loc[(((data['Publisher'].isin(top3_publisher))&(data['Platform'].isin(top3_plateform))&(data['Publisher'].isin(top3_publisher))&(data['Genre'].isin(top3_genre))))]
top3_data
```

```python
# Compare top3 platform these publisher used?
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Publisher',hue='Platform',data=top3_data)
```

```python
# Total value of each publisher
sns.countplot(x='Publisher',data=top3_data)
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Publisher',hue='Platform',data=top3_data,dodge=False)
```

```python
# Stacked bar chart only help in Total
# but if you want to compare use dodge chart
```
